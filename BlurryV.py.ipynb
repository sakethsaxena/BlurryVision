```python
# Importing packages 
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('data/starry_starry_night.jpg')
```

```python
# cv2 loads image as BGR, matplotlib as RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# Applying a gaussian filter - reduing noise 
blur = cv2.GaussianBlur(img,(91,91),0)
plt.imshow(blur)
plt.show()
```

```python
# Attempt 1 - canny edge detection
edges = cv2.Canny(img,100,200)
plt.imshow(edges)
plt.show()

# Kinda failed lol
```

```python
# Attempt 2 - tracing contours in the blurred image
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_1 = img
ret,thresh = cv2.threshold(imgray,127,255,0)
contours, h = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
height, width = img_1.shape[:2]
blank_image = np.zeros((height,width,3), np.uint8)

img_1 = cv2.drawContours(blank_image, contours, -1, (0,255,0), 3)
plt.imshow(img_1)
plt.show()
# this kinda worked
```
